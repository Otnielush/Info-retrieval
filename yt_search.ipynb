```python
import pytube
import pandas as pd
import numpy as np
import youtubesearchpython  as ys

# https://python-pytube.readthedocs.io/en/latest/api.html#module-pytube.extract
# https://pypi.org/project/youtube-search-python/
```

```python
search = 'metallica'
videosSearch = ys.VideosSearch(search, limit = 10)
print(videosSearch.result())
```

```python
res = videosSearch.result()
```
